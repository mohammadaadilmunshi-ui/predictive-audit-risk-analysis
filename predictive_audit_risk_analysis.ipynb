## 2.1 Classification Analysis (25)

**2.1 (25) Perform (at least) 4 types of classification analysis demonstrated in the course.**

**NOTE: Do not include the numerical target variable as a predictor variable in your models.**

```python
X = data.drop(columns=['Income', 'Audit_Type'])
y = data['Audit_Type']
```

```python
train_X, test_X, train_y, test_y = train_test_split(X,y, test_size=0.25, random_state=123)
```

```python
logreg = LogisticRegression(solver = 'liblinear', random_state=123, class_weight = 'balanced')
logreg.fit(train_X, train_y)
logred_pred = logreg.predict(test_X)
```

```python
cnb = ComplementNB()
cnb.fit(train_X, train_y)
cnb_pred = cnb.predict(test_X)
```

```python
dt = DecisionTreeClassifier(random_state=123, class_weight = 'balanced')
dt.fit(train_X, train_y)
dt_pred = dt.predict(test_X)
```

```python
rf = RandomForestClassifier(random_state=123, class_weight = 'balanced')
rf.fit(train_X, train_y)
rf_pred = rf.predict(test_X)
```

## 2.2 Model Evaluation (15)

**2.2 A (8) Evaluate the performance of all models.**

```python
print("Logistic Regression:")
print(classification_report(test_y, logred_pred))

print("Complement Naive Bayes:")
print(classification_report(test_y, cnb_pred))

print("Decision Tree:")
print(classification_report(test_y, dt_pred))

print("Random Forest:")
print(classification_report(test_y, rf_pred))
```

**2.2 B (7) Based on your output in 2.2A, discuss and compare the performance for each of the classification models. Which model is preferred? Why? Explain.**

Based on the output, there are clear differences between each of the models. First, we can look at the recall of each classification model for Audit_Type (class 1). For the Logistic Regression model, there is a recall of 0.81, which means that 81% of the individuals who should be audited were correctly identified. In comparison, the Complement Naive Bayes model has a recall of 0.10, which is very low and would not be useful for businesses trying to identify who should be audited. The Decision Tree model has a recall of 0.52, and the Random Forest model has a recall of 0.53, which are both moderate but still lower than the Logistic Regression model. Overall, since the goal of the business problem is to correctly identify individuals who should be audited, recall is most the important measure to focus on. Based on this, the Logistic Regression model would be the best model to use, since it has the highest recall and is more effective at identifying who should be audited.
